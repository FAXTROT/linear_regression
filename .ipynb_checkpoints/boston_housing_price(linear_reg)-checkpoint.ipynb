```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.data import boston_housing_data
```

```python
X, Y = boston_housing_data()
columns = ['CRIM per capita crime rate by town',
          'ZN proportion of residential land zoned for lots over 25,000 sq.ft.',
         'INDUS proportion of non-retail business acres per town',
         'CHAS Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)',
         'NOX nitric oxides concentration (parts per 10 million)',
         'RM average number of rooms per dwelling',
          'AGE proportion of owner-occupied units built prior to 1940',
          'DIS weighted distances to five Boston employment centres',
          'RAD index of accessibility to radial highways',
          'TAX full-value property-tax rate per $10,000',
          'PTRATIO pupil-teacher ratio by town',
          'B 1000(Bk - 0.63)^2 where Bk is the prop. of b. by town',
          'LSTAT % lower status of the population',
         ]

print(X.shape)
print(Y.shape)
```

```python
df_full = pd.DataFrame(np.ones([506,1]), columns=['Bias']).join(pd.DataFrame(X, columns=columns))
df_full = df_full.join(pd.DataFrame(Y, columns=['Price',]))
#df_full = df_full.join(pd.DataFrame(np.ones([506,1]), columns=['Bias']))
#df_full = df_full.sample(frac=1)
df_full
```

```python
# pre-proccessing

# min-max scalers
column = 'B 1000(Bk - 0.63)^2 where Bk is the prop. of b. by town'
s = df_full[column]
df_full[column] = (s - s.min())/(s.max() - s.min())

column = 'TAX full-value property-tax rate per $10,000'
s = df_full[column]
df_full[column] = (s - s.min())/(s.max() - s.min())
```

```python
df_train, df_test, df_cv = df_full[:300], df_full[300:403], df_full[403:506]
```

```python
y_train = df_train[['Price']].to_numpy()
x_train = df_train.drop('Price', 1).to_numpy()

y_cv = df_cv[['Price']].to_numpy()
x_cv = df_cv.drop('Price', 1).to_numpy()

y_test = df_test[['Price']].to_numpy()
x_test = df_test.drop('Price', 1).to_numpy()

print(x_train.shape)
print(y_train.shape)
```

```python
# init params
m, n = x_train.shape

w = 2*np.random.random([n, 1])-1
#b = np.ones([1, n])

alpha = 10**(-40)
reg_y = 10**(40)
```

```python
# main loop

epochs = [x for x in range(0, 10)]
cost_train = []
cost_test = []
for epoch in epochs:
    h = np.dot(x_train, w)
    C = np.sum( (h - y_train)**2 )/(2*m)
    print(f"Train cost = {C}")
    cost_train.append(C)
    
    #w = w - alpha * np.sum(np.dot( x_train.T, (h-y_train) ), axis=1)/m
    
    # L2 regularization
    w = (1-alpha*reg_y/m)*w - alpha * np.sum(np.dot( x_train.T, (h-y_train) ), axis=1)/m
    
plt.title("Error")
plt.plot(epochs, cost_train, 'r.')
plt.show()    

#w_ideal = np.dot( np.dot( np.linalg.inv(np.dot(x_train.T, x_train)), x_train.T ), y_train)
w_ideal = np.linalg.inv(x_train.T @ x_train) @ x_train.T @ y_train
h = np.dot(x_train, w)
C = np.sum( (h - y_train)**2 )/(2*m)
print(f"Perfect cost = {C}")
```
